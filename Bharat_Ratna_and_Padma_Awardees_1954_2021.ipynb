```python
import pandas as pd

data = pd.read_csv("padmaawards.csv")

data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
# award winner by name

def award_winner_by_name(name):
  winner = data[data["Name"] == str(name)]

  winner = winner[["Award", "Field","Year"]]
  
  winner = winner.sort_values(by= "Year", ascending = False)

  return winner
```

```python
award_winner_by_name("Dr. Sarvapalli Radhakrishnan")
```

```python
# award winner in a given year

def award_winner_by_year(year):
  winner = data[data["Year"] == int(year)]

  winner = winner[["Name","Award","Field","State/Country"]]

  return winner
```

```python
award_winner_by_year(1999)
```

```python
# award winner by field

def award_winner_by_field(field):
  winner = data[data["Field"] == str(field)]

  winner = winner[["Award", "Name","Year"]]

  winner = winner.sort_values(by="Year", ascending = False)

  return winner
```

```python
award_winner_by_field("Medicine")
```

```python
# award type

def award_winner_type(award):
  winner = data[data["Award"] == str(award)]

  winner = winner[["Name","Field","Year"]]

  winner = winner.sort_values(by="Year", ascending = False)

  return winner
```

```python
award_winner_type("Bharat Ratna")
```
